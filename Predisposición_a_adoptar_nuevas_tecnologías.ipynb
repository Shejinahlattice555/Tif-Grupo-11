```python
from google.colab import files


uploaded = files.upload()
import pandas as pd

file_name = list(uploaded.keys())[0]

data = pd.read_csv(file_name)

print(data.head())
```

```python
print(data.isnull().sum())

data = data.fillna("No disponible")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma para la predisposición a adoptar tecnología
plt.figure(figsize=(8, 6))
sns.histplot(data['LikelinessToAdopt'], kde=True, bins=10, color="blue")
plt.title("Distribución de la Predisposición a Adoptar Tecnología")
plt.xlabel("Predisposición")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Relación entre edad y predisposición
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['LikelinessToAdopt'], alpha=0.5, color="green")
plt.title("Relación entre Edad y Predisposición a Adoptar Tecnología")
plt.xlabel("Edad")
plt.ylabel("Predisposición")
plt.show()
```

```python
# Gráfico de barras para nivel educativo y predisposición promedio
education_adoption = data.groupby("EducationLevel")["LikelinessToAdopt"].mean()

education_adoption.plot(kind="bar", figsize=(8, 6), color="orange")
plt.title("Predisposición Promedio por Nivel Educativo")
plt.xlabel("Nivel Educativo")
plt.ylabel("Predisposición Promedio")
plt.xticks(rotation=45)
plt.show()
```

```python
# Frecuencia de uso de laptops/desktops vs predisposición
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Frequency_LaptopDesktop'], y=data['LikelinessToAdopt'], palette="viridis")
plt.title("Relación entre Uso de Laptops/Desktops y Predisposición")
plt.xlabel("Frecuencia de Uso de Laptops/Desktops")
plt.ylabel("Predisposición")
plt.show()
```
