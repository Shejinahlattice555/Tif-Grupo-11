```python
from google.colab import files


uploaded = files.upload()
import pandas as pd

file_name = list(uploaded.keys())[0]

data = pd.read_csv(file_name)

print(data.head())
```

```python
print(data.isnull().sum())

data = data.fillna("No disponible")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograma para la predisposicin a adoptar tecnologa
plt.figure(figsize=(8, 6))
sns.histplot(data['LikelinessToAdopt'], kde=True, bins=10, color="blue")
plt.title("Distribucin de la Predisposicin a Adoptar Tecnologa")
plt.xlabel("Predisposicin")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Relacin entre edad y predisposicin
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['LikelinessToAdopt'], alpha=0.5, color="green")
plt.title("Relacin entre Edad y Predisposicin a Adoptar Tecnologa")
plt.xlabel("Edad")
plt.ylabel("Predisposicin")
plt.show()
```

```python
# Grfico de barras para nivel educativo y predisposicin promedio
education_adoption = data.groupby("EducationLevel")["LikelinessToAdopt"].mean()

education_adoption.plot(kind="bar", figsize=(8, 6), color="orange")
plt.title("Predisposicin Promedio por Nivel Educativo")
plt.xlabel("Nivel Educativo")
plt.ylabel("Predisposicin Promedio")
plt.xticks(rotation=45)
plt.show()
```

```python
# Frecuencia de uso de laptops/desktops vs predisposicin
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Frequency_LaptopDesktop'], y=data['LikelinessToAdopt'], palette="viridis")
plt.title("Relacin entre Uso de Laptops/Desktops y Predisposicin")
plt.xlabel("Frecuencia de Uso de Laptops/Desktops")
plt.ylabel("Predisposicin")
plt.show()
```
